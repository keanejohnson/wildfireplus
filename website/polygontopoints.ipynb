```python
import geopandas as gpd
import numpy as np
```

```python
# read JSON on active fire
url = 'https://opendata.arcgis.com/datasets/5da472c6d27b4b67970acc7b5044c862_0.geojson'
```

```python
datalist = gpd.read_file(url)
```

```python
c = datalist.iloc[:,24]
c
```

```python
def points(poly):
    """ take a polygon in any CRS and return a list of x,y or long,lat tuples"""
    return list(map(tuple,np.asarray(poly.exterior.coords)))

l = points(c[0])
print(l)
```

```python
c1 = c.apply(points)
c1
```
